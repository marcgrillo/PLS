```python
import os   #These two are useful modules
import sys

import numpy as np #Numerical Python
```

```python
# Let's define te directory where you are with the "get current working directory" command...
work_dir = os.getcwd()
print("I am in dir %s" % work_dir)

#We can go up by using the path.dirname
main_dir =os.path.dirname(work_dir)
print("Main directory is %s" % main_dir)

#We can then define the data directory using the os.path.join command
data_dir = os.path.join(main_dir,"Data")
print("data dir is %s" % data_dir)

#..and we can define some output directory with the results
results_dir = os.path.join(main_dir,"results")

#if the directory does not exist, you can create it...
if not os.path.exists(results_dir):
    os.mkdir(results_dir)
    print("Directory %s created !" % results_dir)
```

```python
#legge I dati ordinati dal file dizionario.csv prodotti dallo script FILTER, li scrive su un dizionario
#e stampa i valori chiave
import pandas as pd
colonne=[]
df = pd.read_csv(os.path.join(data_dir,'dizionario.csv'), index_col=0)
d = df.to_dict("split")
diz = dict(zip(d["index"], d["data"]))

print('I valori chiave sono:')
for key, value in diz.items() :
    colonne.append(key)
    print(key)
    
print('\n\nIl numero di studenti è:',len(diz[colonne[0]]))
```

```python
#lettore di riga
riga=1975
for ci in range(len(colonne)):
    print(colonne[ci], '-', diz[colonne[ci]][riga])
```

```python
#Quicklook dello stato degli studenti
att=0
cess=0
sosp=0
for ri in range(len(diz['COD_FIS'])):
    if(diz['STATO_STU_DESC'][ri]=='Attivo'): att=att+1
    if(diz['STATO_STU_DESC'][ri]=='Cessato'): cess=cess+1
    if(diz['STATO_STU_DESC'][ri]=='Sospeso'): sosp=sosp+1
    if(diz['STATO_STU_DESC'][ri]!='Attivo'):
        if(diz['STATO_STU_DESC'][ri]!='Cessato'):
            if(diz['STATO_STU_DESC'][ri]!='Sospeso'): print('riga anomalo:',diz['STATO_STU_DESC'][ri],ri)
            
print('Attivi',att,'\ncessati',cess,'\nsospesi',sosp,'\n somma', att+cess+sosp,'\n totale atteso',len(diz['COD_FIS']))
```

```python
#Un quicklook dei motivi dello stato studente del campione
imm=0
cons=0
rin=0
tr_usc=0
dec=0
altro=0
for ri in range(len(diz['COD_FIS'])):
    if(diz['MOTIVO_STATO_STU_DESC'][ri]=='Immatricolazione'): imm=imm+1
    if(diz['MOTIVO_STATO_STU_DESC'][ri]=='Conseguimento titolo'): cons=cons+1
    if(diz['MOTIVO_STATO_STU_DESC'][ri]=='Rinuncia'): rin=rin+1
    if(diz['MOTIVO_STATO_STU_DESC'][ri]=='Trasferimento in uscita'): tr_usc=tr_usc+1
    if(diz['MOTIVO_STATO_STU_DESC'][ri]=='Decadenza'): dec=dec+1
    if(diz['MOTIVO_STATO_STU_DESC'][ri]!='Decadenza'):
        if(diz['MOTIVO_STATO_STU_DESC'][ri]!='Trasferimento in uscita'):
            if(diz['MOTIVO_STATO_STU_DESC'][ri]!='Immatricolazione'):
                if(diz['MOTIVO_STATO_STU_DESC'][ri]!='Rinuncia'):
                    if(diz['MOTIVO_STATO_STU_DESC'][ri]!='Conseguimento titolo'): altro=altro+1
print('immatricolati=',imm,'\n Conseguimento titolo=',cons,'\n Rinuncia=',rin,'\n Decadenza=',dec,'\n Tresferimenti in uscita=',tr_usc,'\n altro=',altro,'\n somma',imm+cons+rin+tr_usc+dec+altro,'\n dovrebbero essere',len(diz['COD_FIS']))
```

```python
#Scrive i motivi per cui lo studente è Attivo, Cessato o Sospeso (scegliere nello script) 
#e la riga associata allo studente
a='Attivo'
c='Cessato'
s='Sospeso'
analizzo_gli=s

for ri in range(len(diz['COD_FIS'])):
    if(diz['STATO_STU_DESC'][ri]==analizzo_gli): print(diz['MOTIVO_STATO_STU_DESC'][ri], ri)
```

```python
import matplotlib.pyplot as plt
campione=[]
a='Attivo'
c='Cessato'
s='Sospeso'
stato=s
for ri in range(len(diz['COD_FIS'])):
    if diz['STATO_STU_DESC'][ri]==stato : campione.append(diz["MEDIA_VOTO"][ri])
plt.figure(figsize=(10,5))   
plt.hist(campione)
```

```python
import matplotlib.pyplot as plt
i='Immatricolazione'
c='Conseguimento titolo'
r='Rinuncia'
d='Decadenza'
tu='Trasferimento in uscita'
#motivo=i
motivi=[i,c,r,d,tu]
for motivo in motivi:
    campione=[]
    for ri in range(len(diz['COD_FIS'])):
        if diz['MOTIVO_STATO_STU_DESC'][ri]==motivo : campione.append(diz["MEDIA_VOTO"][ri])
    plt.figure(figsize=(10,5))        
    plt.hist(campione, density=False, histtype='bar', label=motivo)
    #plt.hist(campione)
    plt.legend()
```

```python
import matplotlib.pyplot as plt
campione=[]
ctwo=0
cthree=0
cfour=0
cfive=0
csix=0
for ri in range(len(diz['COD_FIS'])):
    if diz['LUNGHEZZA_CARRIERA'][ri]!='NOT DEFINED' : campione.append(diz["LUNGHEZZA_CARRIERA"][ri])
    if diz['LUNGHEZZA_CARRIERA'][ri]==3 : cthree=cthree+1
    if diz['LUNGHEZZA_CARRIERA'][ri]==2 : ctwo=ctwo+1
    if diz['LUNGHEZZA_CARRIERA'][ri]==4 : cfour=cfour+1
    if diz['LUNGHEZZA_CARRIERA'][ri]==5 : cfive=cfive+1
    if diz['LUNGHEZZA_CARRIERA'][ri]==6 : csix=csix+1
plt.figure(figsize=(20,10))
plt.hist(campione, bins=30, density=False, histtype='bar', label='Lunghezza carriera')
#plt.hist(campione)
plt.legend()
print(ctwo,cthree,cfour,cfive,csix)
```
